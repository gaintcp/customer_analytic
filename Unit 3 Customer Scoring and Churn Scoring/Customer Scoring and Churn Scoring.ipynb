```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score, roc_curve, auc

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# Import Data
df = pd.read_csv("Customertravel.csv")
cat_data = pd.DataFrame()

cat_data = pd.concat([cat_data, pd.get_dummies(df['FrequentFlyer'], prefix='FrequentFlyer')], axis=1)
cat_data = pd.concat([cat_data, pd.get_dummies(df['AnnualIncomeClass'], prefix='AnnualIncomeClass')], axis=1)

df['AnnualIncomeClass'] = df['AnnualIncomeClass'].map({'Low Income':0, 'Middle Income':1, 'High Income':2})

cat_data = pd.concat([cat_data, df['AnnualIncomeClass']], axis=1)
X = pd.concat([df[['Age', 'ServicesOpted']], cat_data], axis=1)
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=42)

def fit_evaluate(clf, X_train, X_test, y_train, y_test):
    # fit model to training data
    clf.fit(X_train, y_train)

    # make predictions for train data
    y_pred_train = clf.predict(X_train)

    # make predictions for test data
    y_pred_test = clf.predict(X_test)
    # print evaluation
    print(classification_report(y_test, y_pred_test))
    print('\nConfusion Matrix: \n')
    s = sns.heatmap(confusion_matrix(y_test, y_pred_test), annot=True, fmt='g', cmap='YlGnBu')
    s.set(xlabel='Predicted class', ylabel='True class')
    plt.show()

    #ROC-curve
    fpr_train, tpr_train, _ = roc_curve(y_pred_train,  y_train)
    auc_train = roc_auc_score(y_pred_train, y_train)
    plt.plot(fpr_train,tpr_train, color='Blue', label='train: auc='+f'{auc_train:.2f}')

    fpr_test, tpr_test, _ = roc_curve(y_pred_test,  y_test)
    auc_test = roc_auc_score(y_pred_test, y_test)
    plt.plot(fpr_test,tpr_test, color='Red', label='test: auc='+f'{auc_test:.2f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.legend(loc=4)
    plt.show()

    # Return evaluation metrics as a dictionary
    eval_metrics = {
        'Model': type(clf).__name__,
        'Precision': precision_score(y_test, y_pred_test),
        'Recall': recall_score(y_test, y_pred_test),
        'F1-Score': f1_score(y_test, y_pred_test),
        'Support': len(y_test)
    }

    return eval_metrics

def fit_evaluate_oversampling(clf, X_train, X_test, y_train, y_test):
    ros = RandomOverSampler(random_state=0)
    X_oversampled, y_oversampled = ros.fit_resample(X_train, y_train)

    clf.fit(X_oversampled, y_oversampled)

    y_pred_train = clf.predict(X_oversampled)
    y_pred_test = clf.predict(X_test)

    print("* Evaluation on Test Data (Oversampling) * \n")
    print(classification_report(y_test, y_pred_test))
    print('\nConfusion Matrix: \n')
    s = sns.heatmap(confusion_matrix(y_test, y_pred_test), annot=True, fmt='g', cmap='YlGnBu')
    s.set(xlabel='Predicted class', ylabel='True class')
    plt.show()

    # Return evaluation metrics as a dictionary
    eval_metrics = {
        'Model': type(clf).__name__ + " (Oversampling)",
        'Precision': precision_score(y_test, y_pred_test),
        'Recall': recall_score(y_test, y_pred_test),
        'F1-Score': f1_score(y_test, y_pred_test),
        'Support': len(y_test)
    }

    return eval_metrics

def fit_evaluate_undersampling(clf, X_train, X_test, y_train, y_test):
    rus = RandomUnderSampler(random_state=0)
    X_undersampled, y_undersampled = rus.fit_resample(X_train, y_train)

    clf.fit(X_undersampled, y_undersampled)

    y_pred_train = clf.predict(X_undersampled)
    y_pred_test = clf.predict(X_test)

    print("* Evaluation on Test Data (Undersampling) * \n")
    print(classification_report(y_test, y_pred_test))
    print('\nConfusion Matrix: \n')
    s = sns.heatmap(confusion_matrix(y_test, y_pred_test), annot=True, fmt='g', cmap='YlGnBu')
    s.set(xlabel='Predicted class', ylabel='True class')
    plt.show()

    # Return evaluation metrics as a dictionary
    eval_metrics = {
        'Model': type(clf).__name__ + " (Undersampling)",
        'Precision': precision_score(y_test, y_pred_test),
        'Recall': recall_score(y_test, y_pred_test),
        'F1-Score': f1_score(y_test, y_pred_test),
        'Support': len(y_test)
    }

    return eval_metrics

def run_models_with_sampling(models, X_train, X_test, y_train, y_test):
    result_df = pd.DataFrame(columns=['Model', 'Precision', 'Recall', 'F1-Score', 'Support'])

    for model in models:
        print(f'* {type(model).__name__} * \n')
        normal_eval = fit_evaluate(model, X_train, X_test, y_train, y_test)
        oversampling_eval = fit_evaluate_oversampling(model, X_train, X_test, y_train, y_test)
        undersampling_eval = fit_evaluate_undersampling(model, X_train, X_test, y_train, y_test)

        result_df = result_df.append(normal_eval, ignore_index=True)
        result_df = result_df.append(oversampling_eval, ignore_index=True)
        result_df = result_df.append(undersampling_eval, ignore_index=True)

    return result_df

# Define the models
models = [LogisticRegression(), RandomForestClassifier(), KNeighborsClassifier()]

# Call the function
results = run_models_with_sampling(models, X_train, X_test, y_train, y_test)
print(results)
```
